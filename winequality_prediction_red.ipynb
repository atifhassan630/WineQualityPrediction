```python
import numpy as np 
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns   
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  
from sklearn.linear_model import  LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#read data 
wine = pd.read_csv("winequality-red.csv")
wine.head()
```

```python
#size of data set 
wine.shape
```

```python
#checking the missing values 
wine.isnull().sum()
```

```python
# statistical measure of data 
wine.describe()
```

```python
#number of values for each quality 
sns.catplot(x='quality' , data=wine , kind= 'count' )
```

```python
# volatile acidity vs quality 
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality' , y='volatile acidity' ,data= wine)
```

```python
# citric acid vs quality 
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality' , y='citric acid' ,data= wine)
```

```python
#corelation 
correlation = wine.corr()
```

```python
#heatmap of correlation 
plt.figure(figsize=(10,10))
sns.heatmap(correlation , cbar=True , square=True , fmt= '.1f' , annot = True , annot_kws ={'size':8}, cmap = 'Blues' )
```

```python
#Data preprocessing 
X = wine.drop('quality', axis=1)
X.shape
```

```python
#label binarization 

Y = wine['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
Y.shape
```

```python
# train test data split 
X_train , X_test , Y_train , Y_test  = train_test_split(X , Y , test_size=0.2 , random_state=42)
print(Y.shape , Y_test.shape , Y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=0)
model.fit(X_train , Y_train)
```

```python
#model evaluation 
X_test_prediction = model.predict(X_test)

test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print(f"Accuracy : {test_data_accuracy}")
```

```python
#predictive system 
#8.9,0.62,0.18,3.8,0.176,52,145,0.9986,3.16,0.88,9.2  Bad Quality  5
#7.8,0.58,0.02,2,0.073,9,18,0.9968,3.36,0.57,9.5  Good Quality  7

input_data = (6.8,0.26,0.42,1.7,0.049,41,122,0.993,3.47,0.48,10.5)

#input data to numpy array 
input_data_array = np.asarray(input_data)

#data reshaping 
input_data_reshaping = input_data_array.reshape(1,-1)

prediction = model.predict(input_data_reshaping)
print(prediction)

if prediction == 1 :
    print('Good Quality Wine')
else:
    print('Bad Quality Wine')
```
